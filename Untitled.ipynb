**Histograma**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def histograma(x, f, c):
    fig = plt.figure(figsize=(14, 12), dpi=100)
 
    for i in range(0, c*f):
        plt.subplot(f,c,i+1)
        sns.distplot(x[:,i],bins=10)
        plt.grid()
        if (i==0):
            plt.title("y")
        else:
            plt.title("x _%i" %(i))

histograma(data, 2, 3)
```

**Correlaciones** 

Para llevar a cabo las correlaciones se utilizaron las siguientes instrucciones

```python
def coef_corr(x, y):
    x_1=np.nan_to_num(x)
    cor=np.corrcoef(x_1,y)
    return(cor[0,1])

def corr(x,f,c): 
    fig = plt.figure(figsize=(14, 12), dpi=100)
    for i in range(1, c*f):
        fig.add_subplot(f,c,i)
        plt.scatter(x[: ,i], x[:, 0])
        cor= coef_corr(x[: ,i], x[:, 0] )
        plt.title("Correlaciòn(y, x_%i) = %0.3f" %(i, cor))
        plt.ylabel("y")
    
corr(data, 3, 2)  
```

## Entrenamiento del modelo 

EL modelo requierio de la siguiente instruccion

```python
def modelo(x, y, iterar, error_indi, l_r):
    c_0 = np.array([1., 0.])
    error_c = np.zeros(iterar)
    for i in range(0,iterar):
        y_hat = np.matmul(c_0, np.vstack((x, np.array(np.full((1,len(x)),1)))))
        error_c[i] = 0.5*(np.mean((y_hat-y)**2))
        if (i % error_indi == 0):
            print(" iteracion no, % i"% (i+1),  "error de= \t%0.3f" % (error_c[i]))
        grad_xy  = np.array([ np.mean((y_hat - y)*x), np.mean((y_hat - y)) ])
        c_0 -= l_r * grad_xy
             
    return error_c

x=x_1
y=datos_entrenamiento[:,0]
iterar=150
error_indi=10
l_r=0.0001

# Entrenamiento del modelo

costo_entrenamiento_x1=modelo(x, y, iterar, error_indi, l_r)
```

```python
x=x_2
y=datos_entrenamiento[:,0]
iterar=150
error_indi=10
l_r=0.00000001

# Entrenamiento del modelo

costo_entrenamiento_x2=modelo(x, y, iterar, error_indi, l_r)
```

```python
def predicion(a,b,x_n,x,y):
    ajuste= x_n*a+b
    regr=linear_model.LinearRegression().fit(x[:-292],y[:-292])
    predi=regr.predict(x[-292:])
    
    error_ajuste= 0.5*(np.mean((ajuste-y[-292:])**2)) 
    error_predi= 0.5*(np.mean((predi-y[-292:])**2))
    error_promedio= 0.5*(error_predi+error_ajuste)
    
    return error_ajuste, error_predi, error_promedio

a=24869.04795829
b=3662.26432272
datos_prueba=data[1168:1460,0:6]
x_n= datos_prueba[:,1]   
x=data[:,1].reshape(-1,1)
y=data[:,0]

errores_m1=predicion(a,b,x_n,x,y)
print(errores_m1)
```

```python
a=1.37056538e+02
b=1.10788108e-01
datos_prueba=data[1168:1460,0:6]
x_n= datos_prueba[:,1]   
x=data[:,1].reshape(-1,1)
y=data[:,0]

errores_m2=predicion(a,b,x_n,x,y)
print(errores_m2)
```
